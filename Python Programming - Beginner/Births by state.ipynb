#### Given the following dataset, can you find the top state for baby births? One way to show this visually to a stakeholder would be to plot the births by state (using something like matplotlib).

```python
#import packages 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#reading csv into a dataframe
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')

#viewing dataframe
baby_names.head()
```

```python
#creating dataset with columns that we only need
babynames_states = baby_names[['State', 'Count']]

#group by states
totalbabies_by_states = babynames_states.groupby(['State']).sum().reset_index()

#sorting 
totalbabies_by_states.sort_values('Count',ascending='False')
```
